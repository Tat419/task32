## Основные определения

**Специалисты по анализу данных**, работающие в сфере продаж, часто называются продуктовыми аналитиками.  

Их задача — проанализировать поведение пользователей (покупателей), выявить возможные
проблемы при покупке и предложить решения.   

Например, если рассмотреть этапы взаимодействия  покупателя с интернет-магазином по продаже продукта, то можно четко выделить следующую последовательность:

1. Посещение сайта.  

2. Просмотр карточки товара.  

3. Добавление товара в корзину.  

4. Покупка.  

Как правило, количество пользователей на каждом последующем шаге уменьшается,
и такая ситуация называется **“воронкой”** (воронкой продаж, воронкой конверсии).

Задача  анализа -  понять, на каком
этапе воронки магазин теряет покупателей и какие изменения можно внести, чтобы
улучшить процесс покупок в интернет-магазине.


**Конверсия** — это отношение количества
пользователей на каком-то одном шаге к количеству пользователей на одном
из предыдущих шагов.   

Например, конверсия из посещения  сайта в добавление товара
в корзину рассчитывается так: количество пользователей, добавивших товар в корзину,
делится на количество пользователей, посетивших сайт.


Изучение  воронки  конверсии, которая показывает, как пользователи переходят
от одного шага к другому на сайте - задача продуктового аналитика.

## Проверка гипотез с помощью A/B‑тестов

А/В-тест — это всегда сравнение двух вариантов продукта. Например, двух дизайнов интерфейса или стилей рекламного баннера.   

Когда проводят А/В-тест, аудиторию делят на две группы. Первой (контрольной) показывают старый вариант продукта, второй (тестовой) — новый. После эксперимента берут данные о поведении обеих групп и сравнивают их. Сравнение можно провести на глаз, но лучше **использовать специальные статистические тесты**. Они помогут удостовериться, что разница между поведением пользователей в двух группах — не результат случайного стечения обстоятельств, а объективная реальность.

Название A/B-тест (или А/Б-тестирование) произошло от самого принципа проведения этого эксперимента: сравнения двух версий чего-либо, обозначенных буквами «A» и «B».

В A/B-тесте в один момент тестируется только одно изменение.

Результаты A/B-теста надо проверять на статистическую значимость. Значимости нет — тест не достоверный, ни одна из версий не выиграла. Для статистически значимых тестов вам нужна достаточная выборка.

Надо определиться, какая метрика будет метрикой успеха?   


Обычно  проведение  A/B-теста состоит из следующих действий.
*   Сформулировать гипотезу. Что именно вы хотите проверить и каких
изменений ожидаете?
*  Выбрать метрику, по которой вы планируете понять, правильная ли гипотеза или нет.
*  Создать контрольную и тестируемую версии.
*  Запустить тест.
*  Проанализировать результаты.

В приложенном файле "bank_ab.csv" доступны логи переходов на сайт и последующего оформления карточки за каждый день.

Датасет содержит следующие признаки:  
*   date — дата захода пользователя на сайт
*   id — идентификатор, присвоенный пользователю
*   group — группа теста или контроля
*   converted равно 1, если пользователь оформил договор и 0, если не оформил.

Ваша задача показать графически в какой из групп больше заключенных договоров на протяжении всего периода наблюдения.

Какую долю в каждой из групп составляют заключившие договор от общего числа посетителей в этой группе на протяжении всего периода наблюдения?


### Сначала посмотрим все ли пользователи уникальны?

```python
df.info()
```

```python
df.nunique()
```

Для проведения A/B тестов **важен баланс количества пользователей в тестовой группе и в контрольной**. Убедимся, что пользователей в тестовой и в контрольной группе приблизительно равное количество

```python
df.groupby('group').count()
```

Отличаются ли конверсии в тесте и в контроле?


```python
# конверсия в контроле:
conv_control =  df[(df.group == 'control') & \
 (df.converted == 1)].converted.sum()
seen_control = df[df.group == 'control'].shape[0]
conv_control / seen_control
```

```python
# конверсия в тесте:
conv_test =  df[(df.group == 'test') & (df.converted == 1)].converted.sum()
seen_test = df[df.group == 'test'].shape[0]
conv_test / seen_test
```

Посчитаем отношение конверсий.

```python
# отношение конверсий:
(conv_test/seen_test)/ (conv_control/seen_control ) - 1
```

А можно ли считать это число хорошим исходом? А для проверки значимости этого отличия используем штатные средства анализа.

Вспомним, что  конверсии распределены биномиально (всего 2 варианта значений 0 или 1).

Поэтому **используем  статистический тест для биномиального распределения и проинтерпретируем его результаты**.




Многие тесты, в том числе биномиальный, можно провести с помощью пакета SciPy (Science Python). Это большая библиотека, в которой реализованы статистические методы и научные функции.

Сформулируем следующую **нулевую гипотезу**: «Вероятность конверсии в обеих группах пользователей в нашей выборке ОДИНАКОВА (РАВНА)».

Проверим это с помощью теста.

Подробное описание параметров биномиального теста смотрите https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.binomtest.html

```python
from scipy.stats import binomtest

binomtest(conv_test, seen_test - conv_test, \
          conv_control / (seen_control - conv_control))
```

Смотрим на значение возвращаемого параметра **pvalue** - это и есть искомая вероятность. Она очень близка к нулю. Мы **проверили вероятность получить результат, где нулевая гипотеза верна. Вероятность небольшая. Значит мы можем отвергнуть нулевую гипотезу**.

Важно ещё **проверить достаточно ли долго проводился  тестовый эксперимент**? Если да, то мы могли бы посчитать средние конверсии и сравнить значимость их различия?

Посчитаем количество дней теста, убедимся, что среди них нет пропусков

```python
df['date'].nunique()
```

Посчитаем количество открытых карт и средние конверсии по дням в тесте и в контроле:

```python
res_by_day = df[['group','converted','date']].groupby(['group','date'],
                                                      as_index=False).sum()
res_by_day
```

```python
res_by_day_conv = df[['group','converted','date']].\
groupby(['group','date'],as_index=False).mean()
res_by_day_conv
```

Здесь без визуализации не обойтись. Построим график, отражающий количество оформленных карт по дням

```python
test_n_by_date = list(res_by_day.loc[res_by_day['group']=='test']['converted'])
control_n_by_date = \
list(res_by_day.loc[res_by_day['group']=='control']['converted'])

sns.lineplot(data= test_n_by_date, color = "red", label = 'test')
sns.lineplot(data= control_n_by_date, color ="blue", label = 'control')
plt.ylim(0, 100);
```

Построим график, отражающий конверсию из посещений в оформление карт по дням

```python
test_conv_by_date = \
    list(res_by_day_conv.loc[res_by_day_conv['group']=='test']['converted'])
control_conv_by_date = \
    list(res_by_day_conv.loc[res_by_day_conv['group']=='control']['converted'])

sns.lineplot(data= test_conv_by_date, color = "red", label = 'test')
sns.lineplot(data= control_conv_by_date, color ="blue", label = 'control')
plt.ylim(0, 0.2);
```

Построим распределения средних значений по дням в тесте и в контроле.
На что похоже это распределение?

```python
plt.hist(test_conv_by_date, 20, density=True,color='r',alpha=0.3 );
```

```python
plt.hist(control_conv_by_date, 20, density=True,color='b',alpha=0.6 );
```

Теперь используем знание о том, что с**редние конверсии распределены нормально**.
Используем  статистический тест и проинтерпретируем его результаты.


Подробнее о параметрах теста см. https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html

```python
from scipy.stats import ttest_ind

ttest_ind(test_conv_by_date, control_conv_by_date)
```

Используем тот же статистический тест, чтобы оценить, отличаются ли  результаты для контрольной группы?



```python
ttest_ind(test_n_by_date, control_n_by_date)
```

Получаем, что по результам теста **ttest** средние по дням конверсии отличаются значимо

**Вывод**. Конверсия из переходов на обновленную страницу в оформление дебетовой карты возрастает. И это не случайное совпадение.
